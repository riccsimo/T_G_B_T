```python
import matplotlib.pyplot as plt
from ipywidgets import *
import numpy as np
import pandas as pd
import sys
import taurex.log
taurex.log.disableLogging()
from taurex.cache import OpacityCache,CIACache
%matplotlib widget
```

```python
file_path = "WASP-39b_spectra.csv"

spectra = pd.read_csv(file_path).values

plt.figure(figsize=(10,8))
plt.errorbar(spectra[:,0], spectra[:,1], spectra[:,2], fmt='o', color='blue', markersize=2,
    ecolor='lightblue', elinewidth=1, capsize=2, label="WASP 39b Trasmission Spectrum")
plt.xscale("log")
plt.legend()
plt.xlabel("Wavelength ($\mu$m)")
plt.ylabel("$(R_p/R_s)^2$")
plt.savefig("WASP-39b_spectrum_assignment3.png")
plt.show()
```

```python
spectrum_data = np.column_stack((spectra[:,0], spectra[:,1]/100, spectra[:,2]/100))
np.savetxt('WASP-39b_spectrum_assignment3_taskB.dat', spectrum_data, header="wavelength(micron) (rp/rs)^2 std((rp/rs)^2)",fmt="%.8e")
```
